**Endi navbat matinli ustunlar bilan ishlaymiz**  
Matinli ustunlar bilan ishlashda biz Label Encoding yoki One-Hot Encoding ishlatishimiz kerak.  

**Label Encoding** -- bu ustundagi har bir unique qiymatni raqamlar bilan almashtiradi.  
Bu usul bizning holatga mos keladimi. To'liq emas! Nega?  
Sabab bir necha unique qiymatlarni Label Encode qilganimizda (p, q, s -- 0, 1, 2) model buni 2 0 dan ko'ra muhum deb o'ylaydi.  
Shu sabab biz bu usulni faqat ikkita unique categoryaga ega ustunlarga qo'llaymiz.

**One-Hot Encoding** -- Bu usul aynan categoriyalar bir biridan ustun bo'lmagan bir necha unique categorylar bor holat uchun.
Lekin bu usulda agar unique qiymatlar soni juda muhum. Shu sabab biz buni tekshirishimiz kerak.

```python
def unique_values(col):
    if len(col.unique()) > 2:
        return len(col.unique())
    else:
        return 0
```

```python
unique_vals = df.apply(unique_values)
unique_vals
```

```python
# Unique qiymatlari 2 tada ko'p va kam bo'lgan ustunlarni ajratib olamiz
zero_cols = unique_vals[unique_vals==0].index
cols = unique_vals[unique_vals != 0].index
```

**Label Encoder**

```python
label_encoder = LabelEncoder()

df[zero_cols] = df[zero_cols].apply(label_encoder.fit_transform)
```

**One-Hot Encoder**

```python
encoder = OneHotEncoder(sparse_output=False)

encoded_data = encoder.fit_transform(df[cols])

df_encoded = pd.DataFrame(encoded_data, columns = encoder.get_feature_names_out(cols))

df_encoded.head()
```

Raqamlar bilan almashtirlgan ustunlarni tashlab yuboramiz

```python
df = df.drop(cols, axis=1)
```

Ikkala datasetni yakuniy datasetga birlashtiramiz

```python
df_final = pd.concat([df, df_encoded], axis=1)
```

Yakuniy datasetni saqlab olamiz

```python
df_final.to_csv("../data/processed/mushrooms_cleaned.csv")
```
