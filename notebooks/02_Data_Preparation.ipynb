# Mushrooms Classification
### Data Preparation

Kerakli kutubxonalarni chaqirib olamiz

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv("../data/processed/mushrooms_cleaned.csv", index_col = 0)
df.head()
```

```python
# NaN qiymatlar bor ustunni va foydasiz deb qaror qilgan ustunlarimizni tashlab yuboramiz
df.drop(['stalk-root', 'veil-type', 'gill-attachment', 'veil-color'], axis=1, inplace=True)
```

### Label Encoder and One-Hot Encoder

Endi biz ikki categorydan iborat ustunlarga Label Encoder  
Ikkitadan ko'p categorylarga ega bo'lgan ustunlarga One-Hot Encoderni qo'llaymiz

```python
# Har bir ustundagi categorylar sonini ko'ramiz
group_category = df.nunique()
```

```python
# Ikkita categoryali utunlarni alohida qolganlarini alohidaga ajratamiz
col_two = group_category[group_category <= 2].index
col_other = group_category[group_category > 2].index
```

```python
# Label Encoder ni qo'llaymiz
label_encoder = LabelEncoder()
df[col_two] = df[col_two].apply(label_encoder.fit_transform)
```

```python
# One-Hot Encoder ni qo'llaymiz
onehot_encoder = OneHotEncoder(sparse_output=False)
encoded_data = onehot_encoder.fit_transform(df[col_other])

# Natijani DataFrame ko'rinishiga keltirish
encoded_df = pd.DataFrame(encoded_data, columns=onehot_encoder.get_feature_names_out())
```

```python
df_final = pd.concat([df.drop(col_other, axis=1), encoded_df], axis=1)
```

```python
df_final.to_csv("../data/processed/mushrooms_final.csv", index = False)
```
